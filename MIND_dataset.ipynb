The system should follow a content-based filtering approach where, given a source article (NewsID), it recommends 5 articles based on the logic "Because you liked X, you may like Y".

The implementation must include:
1.  **Feature Engineering**: Utilize 'Category', 'SubCategory', and text features from 'Title' and 'Abstract' in `news_df`.
2.  **Similarity Engine**: Compute item embeddings using TF-IDF vectors on text or by leveraging the provided `entity_embedding.vec` and `relation_embedding.vec` to calculate Cosine Similarity.
3.  **Diversity Mechanism**: Implement a retrieval logic (e.g., Maximal Marginal Relevance) to return 5 items, ensuring a balance between 3-4 highly similar articles and 1-2 peripheral articles from related SubCategories to encourage discovery.
4.  **Inference Pipeline**: Create a function that takes a NewsID as input and returns the top 5 recommendations using a pre-computed similarity index (or an ANN index like FAISS).
5.  **Validation**: Use `behavior_df` to verify if the recommended items appear in the same user sessions or click histories for a sample of users.

## Define Recommendation Strategy

### Subtask:
Outline the Item-to-Item Content-Based Filtering strategy and map out how features from news_df will be used to derive recommendations.


### Recommendation Strategy: Item-to-Item Content-Based Filtering

#### 1. Core Logic
The **Item-to-Item Content-Based Filtering** approach identifies similar articles based on their inherent attributes rather than user-item interaction history. By calculating the similarity between items, we can recommend news that shares common characteristics with articles a user has previously engaged with.

#### 2. Feature Mapping from `news_df`
We will leverage the following features to compute similarity:
- **Category & SubCategory:** Provide high-level thematic alignment (e.g., both articles are about 'sports/football_nfl').
- **Title:** The primary textual representation of the article's core topic.
- **Abstract:** Provides deeper context and detailed keywords not present in the title.

#### 3. Semantic Context via Entities
- **TitleEntities & AbstractEntities:** These fields contain linked entities (e.g., specific people, organizations, or locations). By using these, we can move beyond simple keyword matching and calculate similarity based on shared entities (e.g., both articles mention 'Tim Burton'), providing a more accurate semantic representation of the content.

#### 4. 'Because you liked X, you may like Y' Logic
For a given article **X**, the system will identify a candidate article **Y** if:
- **Feature Overlap:** Y belongs to the same `Category` or `SubCategory` as X.
- **Embedding Proximity:** The textual features (Title/Abstract) of Y have a high cosine similarity score compared to X, potentially using the provided `entity_embedding` to represent the entities found within the text.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# 1. Process textual features: Concatenate Title and Abstract, handle NaNs
news_df['Title'] = news_df['Title'].fillna('')
news_df['Abstract'] = news_df['Abstract'].fillna('')
news_df['CombinedText'] = news_df['Title'] + " " + news_df['Abstract']

# 2. Transform text into TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(news_df['CombinedText'])

print(f'TF-IDF matrix shape: {tfidf_matrix.shape}')
news_df[['NewsID', 'CombinedText']].head()
```

<h3 style="color: #4ed0ceff; font-weight: bold;"> DATA PREPROCESS </h3>

```python
# --- STEP: Data Cleaning ---
print(f"Initial shape: {news_df.shape}")

# 1. Handle missing values
# NAML requires text for Title and Abstract to generate embeddings properly.
news_df['Abstract'] = news_df['Abstract'].fillna('')
news_df['Title'] = news_df['Title'].fillna('')

# 2. Basic text normalization: Lowercasing and stripping whitespace
news_df['Title'] = news_df['Title'].str.lower().str.strip()
news_df['Abstract'] = news_df['Abstract'].str.lower().str.strip()

# 3. Categorical cleanup: Ensure categories are stripped of any trailing whitespace
news_df['Category'] = news_df['Category'].str.strip()
news_df['SubCategory'] = news_df['SubCategory'].str.strip()

# 4. Remove duplicates (by NewsID)
# Sometimes the MIND dataset contains duplicate IDs with different metadata. 
# We keep the first one to ensure a 1:1 mapping for our recommendation matrix.
news_df = news_df.drop_duplicates(subset=['NewsID'])

print(f"Final shape after cleaning: {news_df.shape}")
display(news_df.head())
```

<h3 style="color: #4ed0ceff; font-weight: bold;"> TOKENIZATION </h3>

```python
import collections
import re

# --- STEP: Build Vocabularies ---

def build_dict(data, min_count=1):
    """Builds a dictionary of word counts."""
    counter = collections.Counter()
    for text in data:
        # Simple regex tokenizer (can be replaced by nltk later)
        words = re.findall(r'\w+', text.lower())
        counter.update(words)
    
    # Filter by minimum frequency
    vocab = [word for word, count in counter.items() if count >= min_count]
    
    # Map word -> ID (0 is PAD, 1 is UNK)
    word2idx = {word: (i + 2) for i, word in enumerate(vocab)}
    word2idx['<PAD>'] = 0
    word2idx['<UNK>'] = 1
    return word2idx

# 1. Build Word Dictionary from both Titles and Abstracts
all_text = news_df['Title'].tolist() + news_df['Abstract'].tolist()
word_dict = build_dict(all_text, min_count=2)

# 2. Build Category and SubCategory Dictionaries
def build_cat_dict(series):
    unique_cats = series.unique()
    return {cat: i for i, cat in enumerate(unique_cats)}

category_dict = build_cat_dict(news_df['Category'])
subcategory_dict = build_cat_dict(news_df['SubCategory'])

print(f"Vocabulary Size: {len(word_dict)}")
print(f"Categories: {len(category_dict)}, SubCategories: {len(subcategory_dict)}")
```

```python
# Additional reading
# NAML Paper: "Neural News Recommendation with Attentive Multi-View Learning" (Wu et al., IJCAI 2019).
# MIND Dataset Paper: "MIND: A Large-scale Dataset for News Recommendation" (Wu et al., ACL 2020).

# --- STEP: Tokenization & Sequencing ---
MAX_TITLE_LENGTH = 20
MAX_ABSTRACT_LENGTH = 50
def tokenize_text(text, word_dict, max_len):
    """Converts string to a sequence of word IDs with padding/truncation."""
    words = re.findall(r'\w+', str(text).lower())
    # Convert to IDs, using index 1 ([UNK]) for unseen words
    seq = [word_dict.get(w, 1) for w in words]
    
    # Truncate or Pad with 0 (<PAD>)
    if len(seq) < max_len:
        seq = seq + [0] * (max_len - len(seq))
    else:
        seq = seq[:max_len]
    return seq
# Apply transformations
news_df['title_ids'] = news_df['Title'].apply(lambda x: tokenize_text(x, word_dict, MAX_TITLE_LENGTH))
news_df['abstract_ids'] = news_df['Abstract'].apply(lambda x: tokenize_text(x, word_dict, MAX_ABSTRACT_LENGTH))
# Map Categories to IDs
news_df['category_id'] = news_df['Category'].map(category_dict)
news_df['subcategory_id'] = news_df['SubCategory'].map(subcategory_dict)
print("Preprocessing complete. Example encoded news item:")
display(news_df[['NewsID', 'title_ids', 'abstract_ids', 'category_id']].head())
```
