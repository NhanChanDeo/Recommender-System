# Task
Build an Item-to-Item news recommendation system using the MIND dataset located at "./MINDsmall_train_data/MINDsmall_train/". The system should follow a content-based filtering approach where, given a source article (NewsID), it recommends 5 articles based on the logic "Because you liked X, you may like Y".

The implementation must include:
1.  **Feature Engineering**: Utilize 'Category', 'SubCategory', and text features from 'Title' and 'Abstract' in `news_df`.
2.  **Similarity Engine**: Compute item embeddings using TF-IDF vectors on text or by leveraging the provided `entity_embedding.vec` and `relation_embedding.vec` to calculate Cosine Similarity.
3.  **Diversity Mechanism**: Implement a retrieval logic (e.g., Maximal Marginal Relevance) to return 5 items, ensuring a balance between 3-4 highly similar articles and 1-2 peripheral articles from related SubCategories to encourage discovery.
4.  **Inference Pipeline**: Create a function that takes a NewsID as input and returns the top 5 recommendations using a pre-computed similarity index (or an ANN index like FAISS).
5.  **Validation**: Use `behavior_df` to verify if the recommended items appear in the same user sessions or click histories for a sample of users.

## Define Recommendation Strategy

### Subtask:
Outline the Item-to-Item Content-Based Filtering strategy and map out how features from news_df will be used to derive recommendations.


### Recommendation Strategy: Item-to-Item Content-Based Filtering

#### 1. Core Logic
The **Item-to-Item Content-Based Filtering** approach identifies similar articles based on their inherent attributes rather than user-item interaction history. By calculating the similarity between items, we can recommend news that shares common characteristics with articles a user has previously engaged with.

#### 2. Feature Mapping from `news_df`
We will leverage the following features to compute similarity:
- **Category & SubCategory:** Provide high-level thematic alignment (e.g., both articles are about 'sports/football_nfl').
- **Title:** The primary textual representation of the article's core topic.
- **Abstract:** Provides deeper context and detailed keywords not present in the title.

#### 3. Semantic Context via Entities
- **TitleEntities & AbstractEntities:** These fields contain linked entities (e.g., specific people, organizations, or locations). By using these, we can move beyond simple keyword matching and calculate similarity based on shared entities (e.g., both articles mention 'Tim Burton'), providing a more accurate semantic representation of the content.

#### 4. 'Because you liked X, you may like Y' Logic
For a given article **X**, the system will identify a candidate article **Y** if:
- **Feature Overlap:** Y belongs to the same `Category` or `SubCategory` as X.
- **Embedding Proximity:** The textual features (Title/Abstract) of Y have a high cosine similarity score compared to X, potentially using the provided `entity_embedding` to represent the entities found within the text.

## Algorithm Selection for Item Similarity

### Subtask:
Develop the logic to compute item embeddings and calculate similarity scores between news articles.


**Reasoning**:
I will start by processing the textual features from news_df, concatenating Title and Abstract while handling missing values, and then vectorizing the text using TF-IDF.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# 1. Process textual features: Concatenate Title and Abstract, handle NaNs
news_df['Title'] = news_df['Title'].fillna('')
news_df['Abstract'] = news_df['Abstract'].fillna('')
news_df['CombinedText'] = news_df['Title'] + " " + news_df['Abstract']

# 2. Transform text into TF-IDF matrix
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(news_df['CombinedText'])

print(f'TF-IDF matrix shape: {tfidf_matrix.shape}')
news_df[['NewsID', 'CombinedText']].head()
```

**Reasoning**:
I will now develop the method to represent each article as a weighted vector based on the entity embeddings from entity_df and the entities found in the 'TitleEntities' and 'AbstractEntities' columns.



```python
import json

# Prepare entity embeddings dictionary for fast lookup
# entity_df has ID in column 0 and embeddings in columns 1 to 100
entity_embeddings_dict = entity_df.set_index(0).T.to_dict('list')
embedding_dim = entity_df.shape[1] - 1

def get_article_entity_embedding(row):
    entities = []
    # Parse TitleEntities and AbstractEntities
    for col in ['TitleEntities', 'AbstractEntities']:
        val = row[col]
        if isinstance(val, str) and val.strip():
            try:
                entity_list = json.loads(val)
                for ent in entity_list:
                    entities.append(ent.get('WikidataId'))
            except json.JSONDecodeError:
                continue

    # Filter valid embeddings
    valid_vectors = [np.array(entity_embeddings_dict[eid]) for eid in entities if eid in entity_embeddings_dict]

    if not valid_vectors:
        return np.zeros(embedding_dim)

    return np.mean(valid_vectors, axis=0)

print("Calculating entity-based embeddings for articles...")
# Apply the function to each row
news_entity_embeddings = np.stack(news_df.apply(get_article_entity_embedding, axis=1).values)

print(f'Entity embedding matrix shape: {news_entity_embeddings.shape}')
```

**Reasoning**:
I will now implement a function to calculate cosine similarity for both TF-IDF and entity-based embeddings, then combine them using a weighted average to generate a final similarity score between a source article and all others.



```python
from sklearn.metrics.pairwise import cosine_similarity

# Handle NaNs in entity embeddings (replace with 0 if any mean calculation resulted in NaN)
news_entity_embeddings = np.nan_to_num(news_entity_embeddings)

def get_hybrid_similarities(news_id, tfidf_mtx, entity_mtx, df, alpha=0.5):
    # Get the index of the source article
    try:
        idx = df[df['NewsID'] == news_id].index[0]
    except IndexError:
        return None

    # 1. Text-based similarity (TF-IDF)
    text_sim = cosine_similarity(tfidf_mtx[idx], tfidf_mtx).flatten()

    # 2. Entity-based similarity
    entity_sim = cosine_similarity(entity_mtx[idx].reshape(1, -1), entity_mtx).flatten()

    # 3. Combined similarity (Weighted Average)
    combined_sim = (alpha * text_sim) + ((1 - alpha) * entity_sim)

    return combined_sim

# Example usage for one NewsID
test_news_id = news_df['NewsID'].iloc[0]
similarities = get_hybrid_similarities(test_news_id, tfidf_matrix, news_entity_embeddings, news_df)
print(f'Similarity scores calculated for {test_news_id}. Shape: {similarities.shape}')
```

## Mechanism for Peripheral Recommendations

### Subtask:
Implement a selection logic to retrieve 5 items for a given NewsID, including a diversity mechanism (MMR) and category-based exploration.


**Reasoning**:
I will implement the MMR-based retrieval function along with category-based exploration to select exactly 5 unique news recommendations for a given NewsID.



```python
def calculate_mmr(similarities, tfidf_mtx, candidate_indices, top_k=3, lambd=0.7):
    """Maximal Marginal Relevance selection."""
    selected_indices = [candidate_indices[0]]
    remaining_indices = candidate_indices[1:]

    while len(selected_indices) < top_k and remaining_indices:
        mmr_scores = []
        for idx in remaining_indices:
            sim_to_query = similarities[idx]
            # Calculate max similarity to already selected items using TF-IDF
            sim_to_selected = cosine_similarity(tfidf_mtx[idx], tfidf_mtx[selected_indices]).max()
            score = lambd * sim_to_query - (1 - lambd) * sim_to_selected
            mmr_scores.append(score)

        best_idx = remaining_indices.pop(np.argmax(mmr_scores))
        selected_indices.append(best_idx)
    return selected_indices

def recommend_news(news_id, df, tfidf_mtx, entity_mtx, top_n_candidates=50):
    # 1. Get hybrid similarities
    sims = get_hybrid_similarities(news_id, tfidf_mtx, entity_mtx, df)
    if sims is absent:
        return []

    source_idx = df[df['NewsID'] == news_id].index[0]
    source_row = df.iloc[source_idx]

    # 2. Get top candidates (excluding source itself)
    candidate_indices = np.argsort(sims)[::-1]
    candidate_indices = [i for i in candidate_indices if i != source_idx][:top_n_candidates]

    # 3. Select 3-4 highly similar articles using MMR
    highly_similar_indices = calculate_mmr(sims, tfidf_mtx, candidate_indices, top_k=3, lambd=0.7)

    # 4. Select 1-2 peripheral articles: same Category, different SubCategory
    peripheral_candidates = df[(df['Category'] == source_row['Category']) &
                               (df['SubCategory'] != source_row['SubCategory'])].index.tolist()
    # Exclude source and already selected
    peripheral_candidates = [i for i in peripheral_candidates if i != source_idx and i not in highly_similar_indices]

    # Sort peripheral by similarity
    if peripheral_candidates:
        peripheral_candidates = sorted(peripheral_candidates, key=lambda x: sims[x], reverse=True)
        peripheral_indices = peripheral_candidates[:2]
    else:
        # Fallback if no specific peripheral found
        peripheral_indices = [i for i in candidate_indices if i not in highly_similar_indices][:2]

    # Combine and ensure exactly 5
    final_indices = highly_similar_indices + peripheral_indices
    return df.iloc[final_indices][['NewsID', 'Title', 'Category', 'SubCategory']]

# Test the function
sample_id = news_df['NewsID'].iloc[55]
print(f"Recommendations for: {news_df[news_df['NewsID']==sample_id]['Title'].values[0]} ({news_df[news_df['NewsID']==sample_id]['SubCategory'].values[0]})\n")
recs = recommend_news(sample_id, news_df, tfidf_matrix, news_entity_embeddings)
display(recs)
```
